## Problem:
The CDF of the GPD is given by:
$$
G_{\xi, \beta}(x) = 
\begin{cases}
    1 - \left(1 + \frac{\xi x}{\beta}\right)^{-1/\xi} & \text{if } \xi \neq 0, \\
    1 - e^{-\frac{x}{\beta}} & \text{if } \xi = 0,  
\end{cases}
$$
where $\xi$ is the shape parameter and $\beta>0$ is the scale parameter.

- (Q) Show that, for any $x\ge 0$, $G_{\xi, \beta}(x) \to G_{0, \beta}(x)$ as $\xi \to 0^+$.

[Proof] 

Let $y = \frac{\xi x}{\beta}$. As $\xi \to 0$, $y \to 0$. Using the limit $\lim_{a \to 0} (1 + a)^{1/a} = e$, we have the conclusion. 

[Problem]
The density function of the GPD is given by:
$$
g_{\xi, \beta}(x) = 
\begin{cases}
    \frac{1}{\beta} \left(1 + \frac{\xi x}{\beta}\right)^{-\frac{1}{\xi} - 1} & \text{if } \xi \neq 0, \\
    \frac{1}{\beta} e^{-\frac{x}{\beta}} & \text{if } \xi = 0,
\end{cases}
$$

Use Python to plot the density function for $(\xi, \beta) = (0.5, 1)$ and $(\xi, \beta) = (-0.5, 1)$.
    

```python
import numpy as np

import matplotlib.pyplot as plt

def gpd_pdf(x, xi, beta):
    if xi == 0:
        return (1 / beta) * np.exp(-x / beta)
    else:
        return (1 / beta) * np.power(1 + xi * x / beta, -1/xi - 1)

x = np.linspace(0, 10, 500)
pdf1 = gpd_pdf(x, 0.8, 1)
x = np.linspace(0, 10, 500)
pdf2 = gpd_pdf(x, -0.1, 1)

plt.plot(x, pdf1, label=r'$\xi=0.8, \beta=1$')
plt.plot(x, pdf2, label=r'$\xi=-0.1, \beta=1$')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('GPD Density Function')
plt.legend()
plt.grid(True)
plt.show()
```
